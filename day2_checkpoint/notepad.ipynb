## Описание

Вы — инженер по качеству в компании, производящей электронные компоненты. Известно, что:

- Каждый компонент имеет вероятность **3%** быть бракованным.
- В одной партии **250 компонентов**.
- Система считается **надёжной**, если в партии **не более 7 бракованных компонентов**.

**Задачи**:

1. Определите тип распределения и его параметры.
2. Рассчитайте вероятность, что партия **надёжна** (т.е. бракованных компонентов ≤ 7).
3. Постройте график распределения и отметьте на нём:
   - Вероятность `P(X ≤ 7)` (надёжность партии).
   - Среднее значение и стандартное отклонение.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom
```

### Задай параметры распределния

```python
# количество испытаний
n = 250
```

```python
# вероятность брака
p = 0.03
```

### Какой тип распределения? Напиши текстом

> Биномиальное

```python
# вычисли среднее значение
distr = binom(250, 0.03)
mean = distr.mean()
print(f"Среднее значение = {mean}")
```

```python
# стандартное отклонение
std_dev = distr.std()
print(f"Стандартное отклонение = {std_dev}")
```

```python
# рассчитай вероятность P(X ≤ 7) с помощью подходящей функции
prob = distr.cdf(7)
print(f"Вероятность P(X ≤ 7) = {prob}")
```

```python
print(f"Среднее число бракованных компонентов: {mean:.2f}")
print(f"Стандартное отклонение: {std_dev:.2f}")
print(f"Вероятность надёжности партии (≤7 бракованных): {prob:.4f} ({prob*100:.2f}%)")
```

### Строим график

Если ввёл корректно все данные, выполни нижний блок и посмотри графики

```python
# График распределения
x = np.arange(0, 20)
pmf = binom.pmf(x, n, p)

plt.figure(figsize=(10, 5))
bars = plt.bar(x, pmf, color="skyblue", edgecolor="black", alpha=0.7)
plt.axvline(x=mean, color="red", linestyle="--", label=f"Среднее = {mean:.1f}")
plt.title(f"Биномиальное распределение (n={n}, p={p})")
plt.xlabel("Число бракованных компонентов")
plt.ylabel("Вероятность")
plt.legend()
plt.grid(True, linestyle="--", alpha=0.5)

# Выделение области P(X ≤ 7)
for i in range(8):
    bars[i].set_color("green")

plt.legend(["Среднее", f"P(X ≤ 7) = {prob:.2f}"])
plt.show()
```
