**3.** Дан закон распределения случайной величины: 

|  2   | 5 | 8    | 11   |
|------|---|------|------|
| 0.46 | ? | 0.11 | 0.14 |

* Найди неизвестную вероятность и объясните, почему ответ именно такой. 

> Вероятность равна 1 - 0.46 - 0.11 - 0.14 = 0.29
>  
> Случайная величина обязана принимать какое-либо значение, т.е. сумма вероятностей всех возможных значений должна равняться 1.

* Визуализируй функцию распределения по полученному в предыдущем пункте результату

```python
import seaborn as sns
import matplotlib.pyplot as plt

import numpy as np
import scipy.stats as st
```

```python
# code

values = [2, 5, 8, 11]
probs = [0.46, 0.29, 0.11, 0.14]
my_dist = st.rv_discrete(values=(values, probs))

plt.style.use('bmh')

plt.bar(values, my_dist.cdf(values))
plt.title('Распределение вероятности случайной величины')
plt.xticks(range(2,12))
plt.xlabel('Значения случайной величины')
plt.ylabel('P(xi <= x)')
;
```

**4.** Мы являемся владельцами розничного магазина и хотим оптимизировать нашу обслуживающую систему, чтобы обеспечить удобство покупателей и эффективность работы магазина. Известно, что в среднем в магазин заходит 3 человека в минуту. Нам бы хотелось оценить все вероятности потока людей, что правильно планировать наши ресурсы

* Что это за распределение?

> Распределение Пуассона

* Найти вероятность того, что за 2 минуты в магазин зайдет не более 1 человека.

```python
av_num = 3

p = st.poisson.cdf(1, 2 * av_num)

print(f"Вероятность = {p}")
```

```python
# code
x = np.array(range(0,15))
y = st.poisson.pmf(x, 2 * av_num)
plt.plot(x, y, color='black')
plt.fill_between(x, y, where=(x < 2))
;
```

**5.** Нами закуплен новый станок для работы по металлу. Известно, что время работы этого станка до первого отказа подчиняется показательному(экспоненциальному) распределению со средним значением 1 год. Отказ этого станка для нас это огромные убытки. 

* Какова вероятность, что до первого отказа пройдет не менее 2 лет?

```python
l = 1
p_2_years = 1 - st.expon.cdf(2, scale=1 / l)
print(f"Вероятность = {round(p_2_years, 3)}")
```

**7.** Рост мальчиков возрастной группы 15 лет - нормально распределённая случайная величина $X$ с параметрами $\mu=161$ см и $\sigma=4$ см.
* Какую долю костюмов для мальчиков, имеющих рост от 152 до 158 см, нужно предусмотреть в объёме производства для данной возрастной группы?

```python
mu = 161
sigma = 4

height_distr = st.norm(mu, sigma)

part = height_distr.cdf(158) - height_distr.cdf(152)
print(f"Доля костюмов равна {round(part,3)}")
```

> Доля костюмов равна 0.214

* Сформулируй правило трёх сигм для случайной величины $X\sim\mathcal{N}$.

> $\sigma$ - среднеквадратичное отклонение нормального распределения,  
> $X$ - среднее значение распределения (мат. ожидание) 
>     
> Тогда: 
> * в диапазон ($X - \sigma; X + \sigma$) попадет 66.7% всех значений случайной величины  
> * в диапазон ($X - 2*\sigma; X + 2*\sigma$) - 95% всех значений  
> * в диапазон ($X - 3*\sigma; X + 3*\sigma$) - 99.7% всех значений.

* Визуализируй следующие элементы: 
    * кривую плотности вероятности для указанных параметров и распределения
    * залей выдели цветом нужный фрагмет диаграммы (примерно это должно выглядеть [так](https://i.stack.imgur.com/c8F3F.png), только с твоими параметрами)

```python
# code
x = np.linspace(mu - 4 * sigma, mu + 4 * sigma, 1000)
y = height_distr.pdf(x)
plt.plot(x, y, color='darkblue')
plt.xlabel('Рост')
plt.fill_between(x, 
                y, 
                where=(x >= 152) & (x <= 158),
                alpha=0.5,
                hatch='xxx',
                facecolor='none',
                edgecolor='darkblue',
                linewidth=1.5,
)
plt.annotate(
        '152 см',
        (152, height_distr.pdf(152)),
        textcoords="offset points",
        xytext=(0, 10),
        ha='right',
        fontsize=10,
)
plt.annotate(
        '158 см',
        (158, height_distr.pdf(158)),
        textcoords="offset points",
        xytext=(0, 10),
        ha='right',
        fontsize=10,
)
;
```

❓От чего зависит форма кривой нормального распределения? 

> От величины $\sigma$, т.е. величины среднеквадратичного отклонения

**8**. Среднее время обслуживания одного клиента составляет 10 минут, а стандартное отклонение равно 2 минутам. Выстроилось очередь из 20 клиентов. Какова вероятность, что среднее время обсулуживание этих клиентов будет меньше 11 минут

`Подсказка: Центральная предельная теорема`

> По ЦПТ распределние средних значений по выборкам из 20 клиентов будет иметь: 
> * среднее равное генеральной совокупности, т.е. 10 минут
> * дисперсию $D = \sigma^2 / n = 0.2$
> * стандартное отклонение $\sigma$<sub>new</sub> $= \sqrt{D}$

```python
sigma_2 = np.sqrt(0.2)
av_time_distr_20 = st.norm(10, np.sqrt(0.2))

p4 = av_time_distr_20.cdf(11)

print(f"Вероятность равна {round(p4, 3)}")
print(f"сигма {sigma_2}")
```
